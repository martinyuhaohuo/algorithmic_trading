# Portfolio Theory

## Risk and Return of Portfolio

### Set Up
- A portfolio is simply a collections of assets
- An asset $k$ has return $\tilde{r}_k$, which is a random variable
- Each asset is characterized by the mean of its return, variances of its return, and covariances with other assets' returns
- Mean returns:

    | Asset        | 1 | 2 | … | n |
    |-------------|---|---|---|---|
    | Mean Return | $\bar{r}_1$ | $\bar{r}_2$ | … | $\bar{r}_n$ |
    
- Variances and covariances (covariance matrix) of returns:

    |      | $r_1$ | $r_2$ | … | $r_n$ |
    |------|-------|-------|---|-------|
    | $r_1$ | $\sigma_1^2$ | $\sigma_{12}$ | … | $\sigma_{1n}$ |
    | $r_2$ | $\sigma_{21}$ | $\sigma_2^2$ | … | $\sigma_{2n}$ |
    | $\vdots$ | $\vdots$ | $\vdots$ | $\ddots$ | $\vdots$ |
    | $r_n$ | $\sigma_{n1}$ | $\sigma_{n2}$ | … | $\sigma_n^2$ |

- Covariance of an asset with itself is its variance

### Portfolio of Two Assets
- A portfolio of two assets is characterized by the value invested in each asset:
    - Let $V_1$ and $V_2$ be the dollar amount invested in asset 1 and 2
    - The total value of the portfolio is $V = V_1 + V_2$
    - Weight on asset 1 is: $w_1 = V_1 / V$
    - Weight on asset 2 is: $w_2 = = V_2 / V$
    - $w_1 + w_2 = 1$
- Expected portfolio return: $$\bar r_p = w_1 \bar r_1 + w_2 \bar r_2$$
- Unexpected portfolio return: $$\tilde r_p = w_1 \tilde r_1 + w_2 \tilde r_2$$
- Variance of return on this portfolio: $$Var(\tilde r_p) = E[(\tilde r_p - \bar r_p)^2] = w_1^2 \sigma_1^2 + w_2^2 \sigma_2^2 + w_1w_2 \sigma_{12}$$

### Portfolio of Multiple Assets
- We now consider the portfolio of n assets, with weight on each asset denoted $w_i$:
    $$\sum_i^{n}w_i = 1$$
- Return on the portfolio:
    $$\tilde{r}_p = w_1 \tilde{r}_1 + w_2 \tilde{r}_2 + \cdots + w_n \tilde{r}_n = \sum_{i=1}^{n} w_i \tilde{r}_i $$
- Expected return on the portfolio:
    $$\bar{r}_p = \mathbb{E}[r_p] = w_1 \bar{r}_1 + w_2 \bar{r}_2 + \cdots + w_n \bar{r}_n = \sum_{i=1}^{n} w_i \bar{r}_i$$
- Variance of return on the portfolio:
    $$\sigma_p^2 = \mathrm{Var}[\tilde{r}_p] = \sum_{i=1}^{n} \sum_{j=1}^{n} w_i w_j \sigma_{ij}$$
    - Where: $\sigma_{ii} = \sigma_i^2$
- Volatility (StD) of portfolio return:
    $$\sigma_p = \sqrt{\mathrm{Var}[\tilde{r}_p]} = \sqrt{\sigma_p^2}$$
    - Volatility does not account only individual variances of return, but also covariances between asset returns
- Summary: in order to calculate return variance of a portfolio, we need:
    - Portfolio weights
    - Individual variances of returns
    - All covariances of returns

## Diversification

### Types of Risk
- Risk comes in two types:
    - Non-systematic risk
    - Systematic risk
- Diversification refers to increase the number of assets hold / balance the weight on each asset
- Diversification in general can reduce the first type of risk but not the second type
- <img src="images/5_1.png" width="800">

### Diversification Reduces Non-systematic Risk
Portoflios with more assets in general have lower volatility than portfolios with limited asset:
- <img src="images/5_2.png" width="800">
Given two assets with the same volatility (StD) of return = 35%, a portfolio that mixes between two assets can induce a volatility < 35%, as long as correlation between returns of two assets is not 1 (perfect correlation)
- <img src="images/5_3.png" width="800">

### Diversification Cannot Reduce Systematic Risk
- Consider equally-weighted portfolio of $n$ assets
- We can derive the variance of portfolio return as:
$$
\begin{align*}
\sigma_p^2 & = \sum_{i=1}^{n} \sum_{j=1}^{n} w_i w_j \sigma_{ij} \\
& = \sum_{i=1}^{n} \left(\frac{1}{n}\right)^2 \sigma_{ii} + \sum_{i=1}^{n} \sum_{j \ne i}^{n} \left(\frac{1}{n}\right)^2 \sigma_{ij} \\
& = \left(\frac{1}{n}\right) \left( \frac{1}{n} \sum_{i=1}^{n} \sigma_i^2 \right) + \left( \frac{n^2 - n}{n^2}\right)\left(\frac{1}{n^2 - n}\sum_{i=1}^{n} \sum_{j \ne i}^{n} \sigma_{ij}\right) \\
& = \left(\frac{1}{n}\right) (\text{average variance}) + \left(\frac{n^2 - n}{n^2}\right) (\text{average covariance})
\end{align*}
$$
- As n becomes very large:
    - Contribution of variance term goes to zero (variance of returns of assets is the source of non-systematic risk)
    - Contribution of covariance terms goes to “average covariance” and is not reducible (covariance between returns of assets is the source of systematic risk)

## Optimal Portfolio Selection
- Given an expected return $\bar r_p$, the portfolio that minimizes risk (measured by StD) is a mean-StD (or mean-variance) frontier portfolio 
- That is, an optimal portfolio under the mean-variance standard solves:
$$
\begin{align*}
\min_{\{w_1, \ldots, w_n\}} & \sigma_p^2 = \sum_{i=1}^{n} \sum_{j=1}^{n} w_i w_j \sigma_{ij} \\
& \text{subject to} \\
& (1) \quad \sum_{i=1}^{n} w_i = 1 \\
& (2) \quad \sum_{i=1}^{n} w_i \bar{r}_i = \bar{r}_p \\
\end{align*}
$$
- Portfolio frontier: the locus of all frontier portfolios in the mean-StD plane is called portfolio frontier
- **Efficient frontier:** the upper part of the portfolio frontier gives efficient frontier portfolios (downward part is dominated in term of expect return)

### Portfolio Frontier with Two Assets
- Assume $\bar r_1 > \bar r_2$ and let $w_1 = w$ and $w_2 = 1 − w$
- Then:
    - Expected return of the portfolio: $\bar{r}_p = w \bar{r}_1 + (1 - w)\bar{r}_2$
    - Variance of return: $\sigma_p^2 = w^2 \sigma_1^2 + (1 - w)^2 \sigma_2^2 + 2w(1 - w)\sigma_{12}$
- For a trageted expected return, there is only one satisfying portfolio weight $w$: 
$$ w = \frac{\bar{r}_p - \bar{r}_2}{\bar{r}_1 - \bar{r}_2} $$
- This portfolio is thus naturally the optimal portfolio and form the portfolio frontier (with varying required expected return)
- The risk (StD of return) is:
$$
\sigma_p =
\sqrt{
\left(
\frac{\bar{r}_p - \bar{r}_2}{\bar{r}_1 - \bar{r}_2}
\right)^2 \sigma_1^2
+
\left(
\frac{\bar{r}_p - \bar{r}_1}{\bar{r}_2 - \bar{r}_1}
\right)^2 \sigma_2^2
+
2
\left(
\frac{\bar{r}_p - \bar{r}_2}{\bar{r}_1 - \bar{r}_2}
\right)
\left(
\frac{\bar{r}_p - \bar{r}_1}{\bar{r}_2 - \bar{r}_1}
\right)
\sigma_{12}
}
$$
- Plot $\bar r_p$ against $\sigma_p$ forms the portfolio frontier in two-assets case
- When short sales are not allowed, we add restrictions $w_1 \ge 0, \quad w_2 \ge 0$:
    - We must have: $\bar{r}_1 \ge \bar{r}_p \ge \bar{r}_2$
    - The portfolio frontier can be plotted as:
    - <img src="images/5_5.png" width="800">
- When short sales are allowed, the restriction is removed:
    - Not necessary that $\bar{r}_1 \ge \bar{r}_p \ge \bar{r}_2$

### Portfolio Frontier with Multiple Assets
- With more than two assets, we need to solve the constrained optimization problem for each possible expected portfolio return level
- We then plot the portfolios that solve the constrained minimization problem at varying level of expected return, forming the portfolio frontier
- <img src="images/5_9.png" width="800">
- Observation: 
    - When more assets are included, the portfolio frontier improves (moves toward upper-left)
    - Leading to lower risk at each given expected return level
    - Intuition behind improvement: since one can choose to ignore the new assets, including them cannot make one worse off

### Portfolio Frontier with A Safe Asset
- When there is a risk-free asset, the frontier portfolios are combinations of:
    - The risk-free asset
    - The tangent portfolio (consisted of only risky assets)
- Consider a frontier portfolio $p$ with weight $x$ on the tangent portfolio $q$, and $1−x$ on the risk-free asset:
    - Return is: $\tilde{r}_p = (1 - x) r_F + x \tilde{r}_q$
    - Expected return: $\bar{r}_p = (1 - x) r_F + x \bar{r}_q$
    - Variance is: $\sigma_p^2 = x^2 \sigma_q^2$
- The set of frontier portfolios in this case is a straight line passing through the risk-free asset and the tangent portfolio
- The straight line is named as Capital Market Line (CML), 
- **Tangent portfolio:**
    - A portoflio of risky assets that achieve highest **Sharpe ratio**
    - The point that a line passing though risk-free asset tangents to the efficient frontier consisted of only risky assets
- **Sharpe ratio:** $$\frac{\bar r_p - r_F}{\sigma_p}$$
- Note that all portfolios on the CML achieve same Sharpe ratio (which is slope of the CML line)
