# Market Microstructure Theory

## First Model: Glosten-Milgrom

### Introduction to Market Microstructure Theory
- Market microstructure theory looks at the game played between market participants
- Market participants: 
    - Usually involve a market maker (liquidity provider) against several traders (liquidity taker)
    - This setting represents a quote-driven market: only market makers post standing orders
- What if we want to look at order-driven market:
    - Add a decision for the trader to become a market maker in the first place of the game

### Adverse Selection Model Set Up
- One of the most famous sequential trade model, developed by Glosten-Milgrom
- Market maker:
    - There is a market maker quotes a bid B and a ask A
    - The market maker faces no inventory bankrupcty concerns (no credit limit)
- Security value:
    - Security traded in the market has value V, which is a random variable with two possible realizations $\bar V$ and $\underline V$, with $\bar V > \underline V$
    - The distribution of V is given by: 
    $$\begin{align*} V \sim  \begin{cases} \overline{V} & \text{with probability } 1 - \delta 
    \\ \underline{V} & \text{with probability } \delta \end{cases} \end{align*}$$
- Trader:
    - There are two types of traders in the market: informed trader and uninformed trader
    - Distribution of trader type: 
    $$\begin{align*} T &\sim \begin{cases} \text{Informed} & \text{w.p. } \mu \\ \text{Uninformed} & \text{w.p. } 1 - \mu \end{cases} \end{align*}$$
    - At each t, one random trader (informed or not informed) take one action S
    $$ \begin{align*}
        \text{Informed traders: } \quad
        S &\sim 
        \begin{cases}
        \text{buy} & \text{if } V = \overline{V} \\
        \text{sell} & \text{if } V = \underline{V}
        \end{cases} \\[1em]
        \end{align*}
    $$
    $$
    \begin{align*}
    \text{Uninformed traders: } \quad
    S &\sim 
    \begin{cases}
    \text{buy} & \text{w.p. } \tfrac{1}{2} \\
    \text{sell} & \text{w.p. } \tfrac{1}{2}
    \end{cases}
    \end{align*}
    $$
    - Trades are of one unit at each time t
    - At time t=0, the informed trader will learn the value of V, the value of V kept unchanged after t=0


### Bid-ask Spread in Adverse Selection Model
- At t=0, the probability of observing buy by a trader (without knowing the trader's type) is:
    $$
    \begin{align*}
    P(\text{buy})
    &= P(\text{buy} \mid \underline{V}) P(\underline{V})
    + P(\text{buy} \mid \overline{V}) P(\overline{V}) \\
    &= \frac{(1 - \mu)\delta + (1 + \mu)(1 - \delta)}{2} \\
    &= \frac{1 + \mu(1 - 2\delta)}{2}.
    \end{align*}
    $$
- At t=0, the probability of observing sell by a trader (without knowing the trader's type) is:
    $$
    \begin{align*}
    P(\text{sell})
    &= P(\text{sell} \mid \underline{V}) P(\underline{V})
    + P(\text{sell} \mid \overline{V}) P(\overline{V}) \\
    &= \frac{(1 + \mu)\delta + (1 - \mu)(1 - \delta)}{2} \\
    &= \frac{1 - \mu(1 - 2\delta)}{2}.
    \end{align*}
    $$
- Hence, by Bayesian rule, if buy order is sent to market maker at t=0, we have:
    $$
    P(\overline{V} \mid \text{buy}) 
    = \frac{P(\text{buy} \mid \overline{V}) P(\overline{V})}{P(\text{buy})} 
    = \frac{(1 - \delta)(1 + \mu)}{1 + \mu(1 - 2\delta)}
    $$
    $$
    P(\underline{V} \mid \text{buy})
    = \frac{P(\text{buy} \mid \underline{V}) P(\underline{V})}{P(\text{buy})}
    = \frac{\delta (1 - \mu)}{1 + \mu(1 - 2\delta)}
    $$
- Similarily, if sell order is sent to market maker at t=0, we have:
    $$
    P(\overline{V} \mid \text{sell})
    = \frac{P(\text{sell} \mid \overline{V}) P(\overline{V})}{P(\text{sell})}
    = \frac{(1 - \delta)(1 - \mu)}{1 - \mu(1 - 2\delta)}
    $$
    $$
    P(\underline{V} \mid \text{sell})
    = \frac{P(\text{sell} \mid \underline{V}) P(\underline{V})}{P(\text{sell})}
    = \frac{\delta (1 + \mu)}{1 - \mu(1 - 2\delta)}
    $$
- If competition forcrs profit of market maker to 0 (before t = 0), the market maker will set bid and ask at t=0 as:
    - $A = E[V \mid \text{buy}]$ and $B = E[V \mid \text{sell}]$
    - That is:
        $$
        \begin{align*}
        A & = \underline{V} P(\underline{V} \mid \text{buy}) + \overline{V} P(\overline{V} \mid \text{buy}) \\
        & = \underline{V} \frac{\delta (1 - \mu)}{1 + \mu(1 - 2\delta)} + \overline{V} \frac{(1 - \delta)(1 + \mu)}{1 + \mu(1 - 2\delta)}
        \end{align*}
        $$
        $$
        \begin{align*}
        B & = \underline{V} P(\underline{V} \mid \text{sell}) + \overline{V} P(\overline{V} \mid \text{sell}) \\
        & = \underline{V} \frac{\delta (1 + \mu)}{1 - \mu(1 - 2\delta)} + \overline{V} \frac{(1 - \delta)(1 - \mu)}{1 - \mu(1 - 2\delta)}
        \end{align*}
        $$
    - The bid and ask spread at t=0 would be:
        $$
        A - B = \frac{4(1 - \delta)\delta \mu (\overline{V} - \underline{V})} {1 - \mu^2 (1 - 2\delta)^2}
        $$
        - It is increasing in $\mu$, which is the proportion of informed trader
        - It is also increasing in $\delta(1-\delta)$, which is the uncertainty about the value of the security

### Adverse Selection Model Simulation
- At t=1, to determine bid, the market maker need to consider $E[V \mid \text{(buy, buy)}]$ if "buy" order is actually received at t=0
- Similarily, to determine ask, the market maker compute $E[V \mid \text{(buy, sell)}]$ if "buy" order is received at t=0
- The bayesian rule is again applied to complete these calculations, and we can also compute for subsequent t
- We can run a simulation with $\mu = 0.3$, $\delta = 0.5$, $\overline{V} = 2$ and $\underline{V} = 1$
- The bid-ask spread shrinks over time non-linearly:
    - The bid and ask price get closer to each other over time since as more order ("signals") are observed, market maker can make more and more accurate guess about V's actual realization using Bayesian rule
    - It is also the adoption of Bayesian rule making the shrinkage of bid-ask spread non-linearly
    - At the end, both bid and ask is very close to actual realization of V (which is only revealed to informed trader) --> this is how the price (and market) can reveal information

### Adverse Selection Model Implication
- Trade prices are a martingale with respect to past trade prices and order flow:
    - That is $E[P_{t+1}|F_t] = P_t$, where $F_t$ includes all trades prices and order flow (buy or sell at each time) up to t
    - Proof: $E[P_{t+1}|F_t] = E[E[V|F_{t+1}]|F_t] = E[V|F_t] = P_t$ (by the tower property, and $F_t ⊆ F_{t+1}$)
    - So the autocovariance of price changes would be zero, the future price is un-predictable using past prices and past order flow
- Yet many people in the past thought that there would be “bid ask” bounce when there exists bid-ask spread:
    - If traders alternate buy/sell randomly, then prices mechanically bounce between bid and ask
    - This creates negative autocovariance in price changes
- Glosten-Milgrom shows that it is wrong: we can have a nontrivial bid-ask spread and still observe unpredictable prices

## Second Model: Kyle

### Set Up of Kyle Model
- Kyle (1985) also purposed an adverse selection model for explanning market micro-structure
- Trader:
    - Informed trader:
        - There is only one informed trader who learns a security's actucal value v
        $$v \sim \mathcal{N}(p_0, \Sigma_0)$$
        - Consider price impact of trading: need to determines the quantity of order $x$, $x>0$ if insider purchase from MM, $x<0$ if insider sell to MM
    - Liquidity (noise) trader:
        - Do not learn the security's actucal value v
        - submit net order quantity u (which is a random variable)
        $$u \sim \mathcal{N}(0, \sigma_u^2) \; \text{and} \; u \perp V$$
- Market maker:
    - There is one single market maker
    - Observes total order quantity $y = x + u$
    - Set a market clearing price p for these orders (trade all of these orders at the unifrom price p, for both selling and buying order)
    - No bid-ask spread, and all trades occur in one period



### Equilibrium of Kyle Model
- Assumption:
    - At equilibrium, the informed trader uses a linear function to map v to order size x, that is: $x = \beta v + \alpha$
        - Where $\beta$ is positive, so that informed trader trade more aggresively if v is large
    - At equilibrium, the market maker set market clearnig price as a linear function of net order y, that is: $p = \lambda y + \mu$
        - Where $\lambda$ is also positive, as the market maker knows that larger net order informs high value V
- Profit of insiders:
    $$\mathbb{E}(\pi) = \mathbb{E}\big((v - \lambda(x + u) - \mu)x\big) = (v - \lambda x - \mu)x$$
    - Note that the expected profit (E wrt u) is influenced by x in two ways:
        - On one hand, higher x means earning a margin for more units
        - On the other hand, higher x leak more information about v to MM, resulting in higher P for purchasing each unit of security, reducing profit
    - The optimal x, which is the x made by insider at equilibrium must solve:
    $$x^* = \arg\max_{x \in \mathbb{R}} \mathbb{E}(\pi) = \frac{v - \mu}{2\lambda} \quad \text{if } \lambda > 0$$
    - Hence, we show that at equilibrium, x is a linear function of v
- Price set by market maker:
    - Under competition and zero profit condition, $p = E[v|y]$, that is:
    $$
    p = \mathbb{E}(v \mid y) = \mathbb{E}(v) + \frac{\operatorname{Cov}(v,y)}{\operatorname{Var}(y)} \big(y - \mathbb{E}(y)\big)
    $$
    - The intuiton behind is about linear regression: thinking about regress v on y
    - We can further shows that:
    $$ \lambda = \frac{\sqrt{\Sigma_0}}{2\sigma_u} $$ and $$ p = \lambda y + \mu$$
    - We show that at equilibrium, p is also a linear function of x
    - $\lambda$ is also called illiquility, which measures the price impact of trading by insiders

### Implication of Kyle Model
- Price impact increases in value uncertainty (variance of value of security), decreases in noise trading (StD of noise order number)
- Insiders are most aggressive when volatility is the highest
- If price impact is reduced (due to higher value uncertainty to MM, higher noise trading), insider is more aggressive on trading
- Prices are still a martingale with respect to own past of prices and order flow in this model
- Note that the linear equilibrium is not the unique equilibrium: there could exist more complex forms of equilibrium

### Repeated over Time
- The basic version of Kyle model assumes that trade takes place in only one period
- In a continous version, we repeat the trading in Kyle model's setting, with a constant v revealed to insiders
- The price setted by MM will gradually converge to true relaization of v
- The intuition is that insider trades such that information slowly trickles into price
- As the price converges to true value from below, there is always an incentive for insider to buy from MM: insiders always buy in the simulation

## Multiple Competing Insiders

### Theory
- Holden and Subrahmanyam (1992) develop a multi-period auction model in which multiple privately informed agents compete
- They show that such setting cause most of the private information to be revealed very rapidly and make price quickly converge to true value (2-20 traders case)
- These results are in contrast Kyle (1985) in which the monopolistic informed trader causes information to be incorporated into prices gradually (1 trader case)

### Experiment
- Information revelation:
    - Bossaerts et.al (2013) showed that when there are:
        - 0 insider, nearly no mis-pricing, all trade prices around the expected value of asset
        - 2 and 6 insiders, most signficant mis-pricing relative to the actual value revealed to insider, gradual decline in mis-pricing over time
        - That is, the market has yet to incorporate nearly any information in the price
        - 10 and 15 insiders, signficant decline in mis-pricing over time, seems consistent with the theory
    - Hence, the theory by Holden and Subrahmanyam (1992) holds only when there are abundant insiders
    - The result rejects that competition between insiders should drive prices immediately toward the fully revealing equilibrium via Bertrand pricing (2-insider case)
    - In turn, in 2 insiders trail, a lot of insiders leave money on the table
    - This is consistent with past work showing subjects may not always compete prices down in Bertrand setting, as some subjects setting prices irrationally
- Volatility:
    - Volatility of price traded is negatively correlated with the number of insiders
    - In the first 30s, volatility is highest in markets with six insiders
    - In the final 30s, volatility is highest in markets with two insiders

## Inventory Model

### Inventory Models: Models without Adverse Selection
Optimal inventory by market makers can generates bid-ask spread even with out adverse selection (private information)because of:
- Inventory costs:
    - Risk of holding asset (price change during holding)
    - Monitoring cost
- Unbalanced portfolio of risk assets (sub-optimal diversification): 
    - Suppose the MM accumulates an unbalanced portoflio due to unsymmetric demand and supply of the asset
    - Then the MM will need to diversify portfolio (mean-variance analysis)
    - Leading to adjustment of bid and ask price to reduce inventory imbalance
- Uncertainty about competing MM’s inventories:
    - This makes it a game of incomplete information, need Bayes-Nash equilibrium
    - Additional risk premium due to strategic uncertainty
