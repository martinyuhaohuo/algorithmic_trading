## Pricing by Arbitrage

### Introduction to Pricing by Arbitrage
- Traditional finance prices assets by looking at expected future payoffs, this makes expectations crucial
- However, there are financial assets that should instead be priced by arbitrage
- For example, futures and forward contracts are priced relative to spot contracts (for immediate delivery) regardless of what people expect the future will bring
- Intuition:
    - Consider about future pricing of skimmed milk
    - Price of skimmed milk would always equal the price of whole milk minus the price of butter 
    - Regardless of how health concept about skimmed milk change in future

## Forwards and Futures

### Forward Contract
- Definition:
    - A forward contract is a commitment to purchase at a future date a given amount of a commodity or an asset at a price agreed on today
    - Forward price: the price fixed now for future exchange
    - Long side/position: the buyer (purchasing unit) designated by the contract
    - Short side/position: the seller (selling unit) designated by the contract
- Features:
    - Traded over the counter (sold by the seller), not by exchanges
    - The contract is custom tailored
    - No money changes hands until maturity
    - Non-trivial counter-party risk (how about the seller fail to deliver)
- Limitations:
    - Illiquidity: cannot trade the contract with others
    - Counter-party risk
    - These limitations are overcomed by **futures**

### Futures Contract
- Definition: 
    - A futures contract is an exchange-traded, standardized, forward-like contract that is marked to the market daily
    - Futures contract can be used to establish a long (or short) position in the underlying commodity/asset
- Features:
    - Standardized contracts: specify underlying commodity or asset, quantity, maturity
    - Traded on exchanges and guaranteed by the clearing house: little counter-party risk
    - Gains/losses settled daily—marked to market: the contract value is updated using the new market price, profits or losses are settled daily
    - Margin account required as collateral to cover losses: 
        - Initial margin: deposit required to open an account
        - Maintenance margin: minimum balance that must be kept; if due to loss settled today balance is below minimum, need to restore balance

### Forward and Futures Prices
- Forward/futures prices are linked to spot prices $S_0$ (the price of the underlying asset today)
- At this moment, we can ignore differences between forwards and futures, and define price of forward $F$ or futures $H$ of maturity $T$ as $F = H$
- This contract provides delivery of the underlying at date T
- Note that another way to get the underlying at date T is buying the underlying now and store it until T
- The difference between cost of buy-and-store from cost of forward/futures are:
    - Cost of storing (for commodities)
    - Benefits from storing:
        - Convenience yield (for commodities)
        - Dividends/interests (for financials)
- By arbitrage, the costs of these two approaches must equal, hence: $$F=H=S_0(1+r_F)^T+\text{net cost of storing over T}$$
- Commodities as underlying:
    - Gold: 
        - Easy to store—negligible cost of storage
        - No dividends or benefits
        - Hence, net cost of storage is 0
        - The price of future or forward is $F=H=S_0(1+r_F)^T$
    - Oil:
        - Costly to store
        - Convenience yield for holding physical commodity: can sell the physical commidity if price increases
        - Hence, net cost of storage not equal to 0
        - Extra: not held for long-term investment (unlike gold), but mostly held for future use
        - Let percentage cost of storgae be $c$ in each period, percentage yield by convenience be $y$ in each period
        - The future price would be: $$F = S_0 [1 + r_F − (y − c)]^T = S_0 [1 + r_F − \hat y ]^T$$
        - **Backwardation**: If $\hat y > r_F$, spot price > future price, and future price decreases in T, usually due to high convenience yield
            - For example, if supply for oil suddenly decline, and if oil is holding physically, it brings convenience of selling/using for production
            - This reduces the cost of holding oil physically, thus, due to principle of no arbitrage, also decreases price of futures
            - Note: so a decline in future price of oil over T, does not mean that the expectation about future oil supply will recover! Instead, it is because we expect there will be further shortage, so the convenience of yeild is very high
        - **Contango**: If $\hat y < r_F$, spot price < future price, and future price increases in T, usually due to high cost of storing physical commidity
- Financials as underlying:
    - No cost to store
    - No convenience yield
    - Dividend or interest on the underlying
    - Let the dividend yield be d, the price of a stock future will be $F = S_0 [1 + r_F − d]^T$
    - Index futures:
        - If the underlying is a portfolio of stock (index), futures prices may react quicker to macro-economic news than the index itself
        - This is because trading in the futures market is easier than trading in cash market if the portoflio involves many stock
        - Hence, index futures are very useful to market makers, investment bankers, stock portfolio managers

### Hedging with Forwards
- Suppose the manager of an oil exploration firm, have just struck oil
- The manager expect that in 5 months time the firm will have 1 million barrels of oil
- The manager are unsure of the future price of oil and would like to hedge the position
- Using a forward contract, the manager could hedge the position by selling forward 1 million barrels of oil, gained a fixed net payoff $F$ 5 months later
- Thus, the hedge is perfect

### Hedging with Futures
- One problem with using forwards to hedge is that they are illiquid
- If after 1 month the firm discover that there is no oil, then the firm no longer need the forward contract
- In fact, holding just the forward contract the firm is now exposed to the risk of oil-price changes
- In this case, the firm would want to unwind the position by buying back the contract
- Given the illiquidity of forward contracts, this may be difficult and expensive
- To avoid problems with illiquid forward markets, the firm may prefer to use futures contracts

## Options

### Determinants of Option Value
- Key factors:
    - price of underlying asset S
    - strike price K
    - time to maturity T
    - interest rate r
    - dividends D
    - volatility of underlying asset σ
- Additional factors that can sometimes influence option value:
    - expected return on the underlying asset
    - additional properties of stock price movements
    - other investors’ attitude toward risk

### Binomial Model:
- The key is finding replicating strategy that gives payoffs identical to those of the call option:
    - The assumption is that market prices are reasonable + no arbitrage principle
    - Then based on the principle that initial cost of the replicating strategy must equal the call price, we can know the call option price
    - Detail refers to the chapter 11 Binomial Option Pricing Model 
- Some key features:
    - Note that in a multi-period binomial model, at any non-terminal node, if call option value < strike price, can execute the option and do not wait
    - However, for a no-dividend asset, it is never optimal to execute the option before maturity, eve if early execution is allowed
- Factors used in calculation:
    - Current stock price of underlying asset
    - Magnitude of possible future changes of stock price – volatility
    - Interest rate
    - Strike price
    - Time to maturity
- Factors not used:
    - Probabilities of stock price outcome (upward and downward movements)
    - My attitude towards risk
- Some definition:
    - Number of shares needed to replicate one call option is called hedge ratio or option delt
- Limitations:
    - Price of underlying can take more than two possible values
    - Trading takes place continuously
    - Solution: devide the length of each period to be arbitrarily small

### Risk-Neutral Pricing
- It is a shortcut of binomial model: it provides same valuation of an option as binomial model
- Deatil refer to the chapter 11 Binomial Option Pricing Model 
- Key point: 
    - The current value of any asset can be expressed as: $$PV(CF) = \frac{(q_u CF_u)(q_d CF_d)}{1 + r}$$
    - Knwoing the current value ($CF$), future payoffs ($CF_u$) and ($CF_d$) of an underlying asset + a risk-free asset allow us to list an system of two equations with two unknows
    - Solve them, we can derive $q_u$ and $q_d$, which are risk-adjusted probabilities
    - With solved $q_u$ and $q_d$, and known $CF_u$ and $CF_d$ of the option, we can solve for the current value of the option
    - That is, the price of a asset (option) equals the expected payoff using the risk-adjusted probabilities, discounted at the risk-free rate
    - This is called the risk-neutral pricing formula
