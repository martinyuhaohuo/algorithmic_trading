# Basic Concepts

## Auctions

### Types of auctions:
- By number/role of participants:
    - Single Sided:
        - One seller, multiple buyers auction (e.g. real estate auctions)
        - One buyer, multiple sellers (e.g. government buy submarine)
    - Double Sided:
        - Multiple buyers and sellers (adoped by most exchanges)
- By type of bidding:
    - Open bid vs Closed bid
    - Ascending vs Descending
- By settlement price:
    - First price, second price, uniform price

### Four Types of Single Sided Auctions:
- **English Auction: Open Ascending Price**
    - Price rises continuously
    - Bidders need to indicate their interest to buy at current price (e.g. raising hands)
    - Or drop out when the prices rises above their value (e.g. lowering their hands)
        - Auction ends when only one bidder remains in bidding
    - Can be extended to multiple objects:
        - If traded objects are complementarities (higher value if purchased together), can use combinatorial auction
        - Such auction allow participants to bid for a combination (bundle) of objects
        - Different participants can bid for different combination / or individual objects
- **Dutch Auction: Open Descending Price**
    - Usually faster than English Auction
    - Price falls continuously
    - The first bidder to indicate interest wins the auction
    - Still in use for US Treasury bills and flower auctions in Netherlands 
- **First Price Sealed Bid Auction**
    - Highest bidder wins and pays the highest bid
    - Bids are submitted in sealed form (e.g., envelopes)
- **Second Price Sealed Bid Auction**
    - Highest bidder wins but pays second highest bid
    - Bids are submitted in sealed form

### Comparision between Four Single-Sided Auctions
- Under private value assumption, Dutch Auction and FP auction have strong strategic equivalence (same strategy space and equilibrium outcome)
- Under same assumption, English Auction and SP auction have weak strategic equivalence (different strategy space but match on equilibrium outcome)
- **Revenue Equivalence Theorem:** With independently and identically distributed private values, risk neutral bidders, no interdependent values, the expected revenue of the four types of single-sided auctions are equivalent
- If bidders are risk adverse, however, Dutch auction generates higher expected revenue
- Why: 
    - Increase in risk aversion do not change dominant stratgey in SP or English auction (bidding the private value is still optimal)
    - In FP or Dutch auction, however, risk adverse bidder will bid higher (reduce risk by trading lower payoff from winning for higher chance of winning)

### Two Types of Double Sided Auctions:
- Continuous auctions
    - Orders are processed for trade continuously
    - e.g. active sessions of most stock exchanges
- Call (or Batch) auctions
    - Orders received over a time period, and all collected orders executed at one instance
    - Often used to determine open and close prices of market (e.g. NYSE)

## Market Structure

### Types of Markets
- **Auction (order driven) markets**
    - All investors can announce prices at which they want to buy or sell via brokers
    - Example: Euronext, Tokyo, Inet, Hong Kong, Toronto Stock Exchange
- **Dealer (quote driven) markets**
    - Market makers (dealers) post quotes
    - Quote compromises:
        - Bid price: the price the market maker is willing to pay for the asset
        - Ask price: the price the market maker is willing to sell the asset
    - Other investors sell or buy the asset from the market maker at dealer quotes
    - Example: Nasdaq before 1997, derivates market
- **Mixed/hybrid markets**
    - As in order driven + specialist or market makers
    - Investors can announce prices at which they want to buy or sell via brokers
    - Market makers can also post quotes
    - Example: Frankfurt, London, NYSE, Nasdaq, Matif, Xetra, Eurex

### Types of Orders
- **Limit buy order:** 
    - placing an order of want to buy X shares at price no larger than P per share
- **Market buy order:** 
    - placing an order of want to buy X shares at any price
- **Limit sell order:** 
    - placing an order of want to sell X shares at P per share or above
- **Market sell order:** 
    - placing an order of want to sell 100 shares, at any price
- In industry there are many more order types

### Liquidity Demand and Supply
- Supply of Liquidity:
    - The limit orders which have been placed, and have not yet been executed, cancelled or revised are collected in the open limit order book
    - These limit orders that are standing ready to trade with new incoming orders supply liquidity
- Demand of Liquidity:
    - When traders place market orders (they demand immediacy), these orders are immediately exected at the most appropraite price
        - This means that a market buy order will be matched with the limit sell order with lowest minimum sell price
        - A market sell order will be matched with the limit buy order with highest maximum buy price
        - Therefore, market orders lead to execution of standing limit orders, they demand liquidity
    - For limit orders:
        - If traders place limit buy orders with price above the one of best limit sell order, it lead to immediate execution
        - If traders place limit sell orders with price below the one of best limit buy order, it again lead to immediate execution
    - In turn, placing market/limit orders that are immediately executed demand liquidity

### Dealer (Quote Driven) Markets
- **Description:**
    - Only certain players have the right to post bid and ask quotes and thus supply liquidity (providing immediatecy to traders)
    - These players are called market makers or dealers
    - Example include: Forex market, OTC, US futures markets
- **Obligations:** 
    - Post quotes, limit on spread between bid and ask price, face minimum on quantity of trade, reporting requirements
- **Privileges:** 
    - Right to post quotes, information about order flow and book, lower or no fees paid to the exchange
- **Revenues:**
    - Dealers try to buy at low price (bid price they set) & sell at high price (ask price they set)
    - Earn bid-ask spread
- **Cost:**
    - There is a time between purchase & sale of the asset
    - During this period, dealer need to hold security in inventory
    - This lead to risk bearing cost/inventory cost
    - For example, if market buy (investor buys security from the dealer) is followed by permanent price increase of the security, dealers bear a loss
    - Especially common if market orders placed by informed agents
- **Competition between Dealers:**
    - Under perfect competition, bid-ask spread should soley reflect compensation for inventory costs & adverse selection costs 
    - e.g. if traders have more information than dealer, market buy happen only if followed by permanent price increase, market sell happen only if followed by price decrease
    - If dealers ahve market power, spreads will reflect rents
- **Benefit of Dealer Market:**
    - There is a fixed cost to provide liquidity: 
        - Ensure connection with electronic system throughout the session (taking liquidity through market order, on the other hand, only require connection at time placing order)
        - Monitor order flow continuously to update quotes, otherwise incur loss (stay on the floor historically)
        - Invest in information acquisition (order flow, news)
        - These cost exist regardless to the trade volume
    - Market marker trade frequently with many investors, thus can efficiently spread this cost over many trades
        - On the other hand, if the market is order driven, investors placing limit order need to incur the fixed cost
        - It is hard to spread the cost, as they do not trade frequently as market maker
    - Finally, it takes initial liquidity to attract orders
        - Market makers post quotes even when order flow is low, this guarantees liquidity at day one of the market (priming the pump)
        - With out market makers, for a new market, there is intially no order, thus attracting no trader, reinforcing the no-order dilemma
- **Drawback of Dealer Market:**
    - There is a risk that dealers collude (increase bid-ask spread)
    - Evidence: 
        - Before 1997, Nasdaq dealers traded only even eighths
        - After a revelaing article, suddenly use odd eighths
    - SEC Order Handling rule (1997) required that limit orders must be displayed
    - Now investors compete more to supply liquidity (spreads decreased)


### Mixed Markets
- Investors placing limit orders along with professional market makers stand ready to serve orders to buy or sell (provide immediacy)
- Example: 
    - in NYSE, investors can enter limit orders in the book
    - in addition, specialists (marker makers of NYSE) manages the book and can enter their quotes
    - in 2006, only 10% orders are traded by specialist in NYSE, 70% are by investors through electrocial system (due to electronic trading)

### Impact of Electronic Trading on Market Micro-structure
- Electronic trading system connect markets, exemplify competition between exchanges
- In USA, this connection is formal: 
    - National Market System connects exchanges 
    - There is also requirement on brokers to give best execution to orders arriving at any node in the system
    - That is, if the best ask price appears in NASDAQ, limit buy order posted on NYSE cannot be executed at a worse price
    - It will be either rounted to NASDAQ (by broker) or waited in the limit book

## Market Session

### A Day at the Exchange
- **Pre-open:**
    - Investors can post limit buy or sell orders to the limit order book
        - All buy orders cumulated to form demand function (discrete)
        - All sell orders cumulated to form supply function (discrete)
    - At the time of openning, a call auction is employed to set the openning price (in NYSE, London Stock Exchange, Paris Bourse, ...)
    - The openning price is set to maximize trading volume
        - Assume list of limit sell orders: price:order unit:cumulated supply
            - [240:10:10, 250:5:15, 260:10:25, 270:5:30, 280:5:35]
        - Assume list of limit buy orders: price:order unit:cumulated demand
            - [240:5:30, 250:5:25, 260:5:20, 270:10:15, 280:5:5]
        - The traded volume at price P can be pinned down as $f(P) = min[supply(P), demand(P)]$
        - It is maximized at P = 260, with traded volume $ min[25, 20] = 20$
    - All limit orders with bid (buy) price > P and ask (sell) price < P are executed (filled) at the uniform price P (openning price)
        - However, because of discrete pricing grid, at the volume maximizing price, supply is not always equal to demand
        - If that happens, some orders offering price = openning price will not be exeucted (in our case, 5 unit of sale with P = 260 remain unexecuted)
        - Which orders are left are based on time pirotiy: first in first out
    - If there are many traders, each will have little impact on openning price: 
        - Price at which limit order is placed mainly influences its execution probability
- **After-open:**
    - Executed buy and sell orders are no longer in the book
    - Unexecuted buy orders (placed at price below the opening price) and sell orders (placed above the clearing price) remain in the book
    - The list of standing limit sell order:
        - [240:0:0, 250:0:0, 260:5:5, 270:5:10, 280:5:15]
    - The list of standing limit buy order: 
        - [240:5:10, 250:5:5, 260:0:0, 270:0:0, 280:0:0]
    - Ask price: best offer to sell, in this case 260
    - Bid price: best bid to buy, in this case 250
    - Bid-Ask spread: difference between best bid & best ask, in this case, 10
    - Depth at best bid: unit of buy at bid price, in this case, 5
    - Depth at best ask: unit of sale at ask price, in this case, 5
    - After market opens, continous double-sided auction is employed to match orders:
        - Investors can place new market or limit orders in the book or cancel limit orders that have not been filled
        - If more than one buy/sell order are present at a given price, time priority applies for execution
        - Double auction: match one buyer and one seller
        - Open: all the traders can see the order book
        - Price: standing price of orders in the book
    - Exact mechansim of order matching:
        - If a new limit order with bid = 265 of buying 10 units is placed, it will be firstly matched with the limit sell order of 5 units at ask = 260
        - 5 units of buy is fullfilled at price = 260
        - broker receiving it cannot execute it at a higher price (e.g. match it with order having ask price = 261 if exist)
    - This is **price priority**, broker must seek best execution for orders
        - price piority is prior to time piority during execution
        - why price priority: 
            - prevent brokers from ripping off customers
            - to incentivize investors to place bids in line with their valuation of the asset
        - price priority easier to enforce if:
            - transparent markets & information widely disseminated
            - centralized computerized order book
    - Note that the limit buy order want to buy 10 units at maximum price 265
        - But the limit sell order with ask price = 260 has only 5 units
        - If there is limit sell order with ask price < 265 >260, it will be matched with left 5 units
        - In our case, there is not, so left 5 units of the limit buy order is standing in the limit order book
        - Now the list of standing limit buy order: [240:5:15, 250:5:10, 260:0:0, 265:5:5, 270:0:0, 280:0:0]
        - Now the list of standing limit sell order: [240:0:0, 250:0:0, 260:0:0, 270:5:5, 280:5:10]
        - Ask price: 270
        - Bid price: 265
        - Bid-Ask spread: 5

### Statistical Regularities due to Double Auction
- Trade volume and price volatility is large after market open (unwinding overnight position + uncertainty about price), decline during the day, peak again due to the open of US market, decline after, and increase again approaching the market close (arrival of macro-economics news)
- Bid-Ask spread change over the day and have some regularities
- Stock price is fat-tailed, extreme price changes occur much more frequently than normal models predict

## Terminology

### Short Selling
- Short selling is selling a unit of an asset that the investor do not currently own
- Investor get the cash for selling it now but will have to pay back the asset in the future
- If investor holds an asset short (negative holdings) at the end of a period and it pays a dividend, the investor will be required to pay that dividend
- Shorting is also usually associated with transaction fees or shorting costs (interest paid for borrowing the asset)
- Shorting limits will be usually shown in **available balance**

### Four things need to be determined for a trade (match of order):
- Who (which orders) trades?
- At what price does trade occur at?
- How many units are traded?
- What happens if there is excess?

### Terms in trade
- **Types of balance:**
    - Settled balance: cash/asset from transations fully settled
    - Avaliable balance: cash/asset can be used for new trade 
    - Cash example:
        - Avaliable cash = settled cash - cash reserved for limit buy orders
    - Asset example:
        - Avaliable asset = settled asset - asset reserved for limit sell orders + shorting limit
- **Market:**
    - A market features an order book (order form) and trade history where you can buy and sell units of an Asset
    - A market has rules such as minimum and maximum unit price, quantity, and tick size (minimum unit of price) for orders
    - These rules may differ between markets
- **Public Market:**
    - A Market where all standing orders and trade history are publicly visible and available to all participants
    - Orders are sent to all participants with no discrimination
- **Private Market:**
    - A Market where orders must be address to a specific participant
    - That order is only visible and available to the specified participant
    - When placing an order in a private market, you will also choose the participant you are sending the order to
- **Marketplace:**
    - A marketplace is a collection of Markets
    - A marketplace may feature both Public and Private Markets at the same time
- **Asset:**
    - The item/good/financial tool you buy or sell in the market
    - May or may not have an associated payoff (liquidating value) to them
- **Holdings:** 
    - How many units of settled asset belong to you
    - Holdings also include shorting limits (e.g. shorting limit = 5, then if shorting 5 units, avalaible asset from 5 to 0, settled asset from 0 to -5)
- **Session:**
    - A trading period (from marketplace open to close)
